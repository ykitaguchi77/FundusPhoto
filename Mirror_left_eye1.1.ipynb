#**Mirror_Left_eye**
左眼の眼底を左右反転するスクリプト

```python
from PIL import Image, ImageOps
import os
import time
from IPython.display import display
import numpy as np
import shutil


#google driveをcolabolatoryにマウント
from google.colab import drive
drive.mount('/content/drive')


orig_path = '/content/drive/MyDrive/Deep_learning/FundusPhoto/cropped_img_2'
dst_path = '/content/drive/MyDrive/Deep_learning/FundusPhoto/flipped_img_2'

#移動先のフォルダを作成
#もしすでにあれば削除
try:
    shutil.rmtree(dst_path)
except:
    pass

if os.path.exists(dst_path):
    shutil.rmtree(dst_path)
os.makedirs(dst_path, exist_ok=True)
```

```python

k=0
for i in os.listdir(orig_path):
    filename = os.path.splitext(i)[0]
    side = filename[-1]
    img = Image.open(orig_path + "/" + i)
    if side == "R":
        img.save(dst_path + "/" + i)
    elif side == "L":
        img_mirror = ImageOps.mirror(img)
        img_mirror.save(dst_path + "/" + i)
    k+=1
    num_images = len(os.listdir(orig_path))
    print(str(k)+"/"+str(num_images)+" images converted")
```

#**ファイル数の確認と画像表示**

```python
def countfile(path):

    # フォルダ内の全ファイル名をリスト化
    files = os.listdir(path)
    
    # リストの長さ（ファイル数）を取得
    count = len(files)
    
    # ファイル数を確認
    print(count)


def showfile(path, num):
    files = os.listdir(path)
    imgPIL = Image.open(path+"/"+files[num])  # 画像読み込み
    display(imgPIL)
```

```python
countfile(orig_path)
countfile(dst_path)
showfile(dst_path, 2)
```
